```python
import requests

from bs4 import BeautifulSoup


TARGET_URL = "https://www.jalan.net/kankou/370000/g2_79/" # スクレイピングしたいURLをここに書く

html = requests.get(TARGET_URL)  # HTMLを取ってくる
soup = BeautifulSoup(html.content, "html.parser")  # HTMLを解析する
title = soup.find("title")  # データを抽出する
```

```python
url = 'https://www.google.co.jp'
driver.get(url)
```

```python
elem_search_word = driver.find_element_by_id("APjFqb")
elem_search_word.send_keys("selenium")

elem_search_word.submit()
```

```python
elem_search_word = driver.find_element_by_id("APjFqb")
elem_search_word.clear()
elem_search_word.send_keys("ポケモン 人気")

elem_search_word.submit()
```

```python
elem_search_hps_list = driver.find_elements_by_class_name("yuRUbf")
elem_path_list = list()

for elem_hp in elem_search_hps_list:
    
    elem_path_list.append(elem_hp.find_element_by_tag_name('a').get_attribute('href'))


elem_path_list
```
